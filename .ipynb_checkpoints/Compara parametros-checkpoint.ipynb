```python
import pandas as pd
import Procedatos as P
import Graficos as G
import seaborn as sns
import numpy as np
```

# Heatmaps

```python
data = P.leer_datos([1,3,6,9],[1,3,6,9],[0],verb=False)
G.graficar_heatmaps(data, parametros=['Memoria','Num_predic'])
```

* El promedio de `Attendance` más bajo es con memoria 1 y con 3,6,9 predictores.
* La mayor desviación estándar en el `Attendance` es con memoria 1 y con 3,6,9 predictores.
* El mayor `Efficiency` es con memoria 9 y con 1 solo predictor.
* El menor `Gini` es con memoria 1 y con 9 predictores.
* El menor `Iniquity` es con memoria 3 y con 9 predictores.

En general:

* Aumentar la memoria disminuye el promedio y la desviación estándar del `Attendance`.
* Aumentar el número de predictores disminuye el promedio del `Attendance`, aunque aumenta su desviación estándar.

# Compara dos modelos

```python
data1 = P.leer_datos([9],[1],[0], verb=False)
data1['Modelo'] = '1-pred'
data2 = P.leer_datos([9],[9],[0], verb=False)
data2['Modelo'] = '9-pred'
data = pd.concat([data1,data2])
G.comparar_desemp(data)
```

# Compara espejos y sin espejos

```python
data1 = P.leer_datos([9],[1],[0], verb=False)
data1['Modelo'] = 'Con espejos'
data2 = P.leer_datos([9],[1],[0], verb=False, espejos=False)
data2['Modelo'] = 'Sin espejos'
data = pd.concat([data1,data2])
G.comparar_desemp(data)
```

Es mejor tener espejos, tanto en desviación estándar del `Attendance` como en el `Efficiency`.
